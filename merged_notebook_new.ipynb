#### 2. Trend Analysis of Calgary's Unemployment Rate


## Guiding Question 2 : 
What is the trend in Calgary's total unemployment rate on a month-to-month basis
and a year-to-year basis?

The code converts the 'Date' column into datetime format and sets it as the index so that time series analysis is facilitated.It computes both the month-to-month (Monthly Change) and year-to-year (Yearly Change) changes in the unemployment rate in Calgary.Finally, the code produces a detailed plot that displays:
The original unemployment rate over time.
The changes month to month.
The changes year to year.
With bold headings, rotated date headings, and small format, the graph manages to reveal trends in the unemployment rates.

```python
df['Date'] = pd.to_datetime(df['Date']) #converts the values in the Date column from a string or another format into Python's datetime objects.
df.set_index('Date', inplace=True) #Date column as the index of the DataFrame.

df['Monthly Change'] = df['Calgary (CER) Unemployment rate (%)'].diff() #.diff calculates the month-to-month change in Calgary’s unemployment rate
#new column called Monthly Change is added
df['Yearly Change'] = df['Calgary (CER) Unemployment rate (%)'].diff(12) #calculates the difference between the current value and the value 12 months prior
#new column named Yearly Change is added

plt.figure(figsize=(12, 6)) #12 inches wide by 6 inches tall)
plt.plot(df.index, df['Calgary (CER) Unemployment rate (%)'], label='Unemployment Rate')
plt.plot(df.index, df['Monthly Change'], label='Monthly Change')
plt.plot(df.index, df['Yearly Change'], label='Yearly Change')
plt.xlabel('Date')
plt.ylabel('Change in Unemployment Rate (%)')
plt.title('Calgary Unemployment Rate Trends')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout() #makes sure fit well within the figure boundaries without overlapping
plt.show()
```

##### Strong Correlations:
## Unemployment rate and yearly change
Observations:
There is a very close connection between the year-over-year change in the unemployment rate and the unemployment rate itself. That is, when the economy is stressed—such as in 2020—both the unemployment rate and the year-over-year change experience extreme spikes. This means that when the unemployment rate is highest, the year-over-year change is large because it reflects the built-up effect over the course of the year.

Interpretation
The tight relationship between the unemployment rate and the change on a year-over-year basis is a reflection that longer-term, cumulative effects of economic dislocation are better reflected on an annual basis. In the periods when the unemployment rate spikes, the year-over-year change is a significant indicator of the degree of economic dislocation.

### 4. Inflation Rate and Retail Sales Analysis

## Guiding Question 3 :

This plot reveals the relationship between inflation rates and retail sales. Areas with more overlapping data points (higher density) are shaded more intensely.Fewer data points have lighter colors. Hexbin plot is chosen because it is helpful in large datasets where scatter plots might appear too crowded.

```python
plt.figure(figsize=(12, 6))

# Create a 2D histogram (hexbin) instead of a scatter plot
#Creates a hexagonal binning plot, which is a type of 2D histogram. It divides the plot area into hexagonal cells and counts the number of data points in each cell.
hb = plt.hexbin(df['Calgary (CMA) Inflation Rate (%change)'],
                df['Calgary (CER) Retail Sales ($billions)'],
                gridsize=50, cmap='YlOrRd') #color map ('Yellow-Orange-Red')
#'gridesize'Sets the number of hexagonal cells. Higher values mean smaller hexagons and more granularity.

plt.colorbar(hb, label='Count')

plt.title('Inflation Rate vs. Retail Sales in Calgary')
plt.xlabel('Inflation Rate (% change)')
plt.ylabel('Retail Sales ($billions)')
plt.show()
```

```python
The goal of this code below is to convert the monthly unemployment data into yearly data by aggregrating relavant columns.
The mean was for wage rate, unemployment rate and inflation rates columns since their metrics show averages over time.
Sum was used for employment changes, building permits, retail sales and housing stats columns because their metrics accumulates monthly total counts over the years. 
This makes visualizing easy to intrepret and captures economic trends over the years. Also working age population data for Calgary was extracted to analyze
the number of people working within the age range 16-65years over the years. According to the City of Calgary, younG persons within the age 15, 16, 17 are permitted 
to work, however with certain restrictions to a number of hours and a level of supervision required (City of Calgary, n.d.). The age column was converted to numeric to ensure
proper filtering.
```

```python
# Converting to numeric
Unemployment_data = Unemployment_data.replace(',', '', regex=True)
extracted_Unemployment_data = Unemployment_data.apply(pd.to_numeric, errors='coerce')
```

```python
# The unemployment rate were monthly, converted to yearly by finding the aggregation of the columns
Unemployment_yearly = extracted_Unemployment_data.groupby('Year').agg({
    "Calgary (CER) Unemployment rate (%)": "mean",
    "Alberta Unemployment Rate (%)" : "mean",
    "Canada Unemployment rate (%)" : "mean",
    "Calgary (CER) Total Employment ('000 Persons)" : "sum",
    "Calgary (CER) Total Employment month-over-month change ('000 Persons)" : "sum",
    "Calgary (CER) Total Employment year-over-year change ('000 Persons)" : "sum",
    "Calgary (CMA) Average Hourly Wage Rate ($)" : "mean",
    "City of Calgary Building Permits Residential Value ($)" : "sum",
    "City of Calgary Building Permits Non-Residential Value ($)" : "sum",
    "Calgary (City) Total Housing Starts (units)" : "sum",
    "Calgary (CMA) Total Housing Starts (units)" : "sum",
    "Calgary (CER) Retail Sales ($billions)" : "sum",
    "Calgary (CMA) Inflation Rate (%change)": "mean"
}).reset_index()
display(Unemployment_yearly) 
```

```python
# Converting age column to numeric and filtering the data for ages 16-65
Population_data = Population_data.rename(columns={'Period':'Year'})
Population_data["Age"] = pd.to_numeric(Population_data["Age"], errors="coerce")
working_age_calgary = Population_data[(Population_data["Age"] >= 16) & (Population_data["Age"] <= 65)]
display(working_age_calgary)
```

```python
# Plotting showing calgary working population over the years
working_age_group = working_age_calgary.groupby(['Year'])['OriginalValue'].sum().reset_index()
working_age_group.rename(columns={'OriginalValue': 'Total_Working_Population'}, inplace=True)
plt.figure(figsize=(8, 6))
plt.plot(working_age_group['Year'], working_age_group['Total_Working_Population'], marker='o', linestyle='-', color='b', label='Working Population')

plt.title('Calgary working population', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('TotalPopulation', fontsize=12)
plt.xticks(working_age_group['Year'], rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

We are using a dataset based on the employment rate of different industries.The below code processes and visualizes industry employment trends over the years. 

```python
#Loading data
industries = "industries_cleaned.csv"
industries_data = pd.read_csv(industries)
display(industries_data.shape)
```

```python
display(industries_data.head())
display(industries_data.dtypes)
```

```python
missing_values = industries_data.isnull().sum()
display(missing_values) #display missing values 
```

```python
industries_data['Date'] = pd.to_datetime(industries_data['Date'], format = '%b-%y') # changing to datetime
#Converts the Date column to datetime format '%b-%y : which means 'Jan-21', 'Feb-20'
industries_data['Year'] = industries_data['Date'].dt.year #Extracting the year
display(industries_data.head()) 
```

```python
# Interested Industries data type converted to numeric data types
convert_columns = ["Construction", "Forestry, fishing, mining, quarrying, oil and gas 10 11", 
                   "Agriculture 9", "Finance, insurance, real estate, rental and leasing", "Educational services"]
#converting to numeric
industries_data[convert_columns] = industries_data[convert_columns].replace(',', '', regex=True).astype(float)
display(industries_data.dtypes)
```

```python
# Performing aggregation for interested industries
#Groups data by Year and sums employment counts for each industry
yearly_industries = industries_data.groupby('Year').agg({"Construction" : "sum", "Forestry, fishing, mining, quarrying, oil and gas 10 11" : "sum",
                                                         "Agriculture 9" : "sum", "Finance, insurance, real estate, rental and leasing" : "sum",
                                                         "Educational services" : "sum"}).reset_index()
display(yearly_industries.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size and style
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Plot each industry's trend over the years
for column in ["Construction", "Forestry, fishing, mining, quarrying, oil and gas 10 11",
               "Agriculture 9", "Finance, insurance, real estate, rental and leasing",
               "Educational services"]:
    plt.plot(yearly_industries["Year"], yearly_industries[column], marker="o", linestyle="-", label=column)

# Customize the graph
plt.xlabel("Year", fontsize=12)
plt.ylabel("Employment Count (Summed Over Year)", fontsize=12)
plt.title("Industry Employment Trends Over the Years", fontsize=14, fontweight="bold")
plt.legend(title="Industries", bbox_to_anchor=(1.05, 1), loc="upper left")  # Move legend outside

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

##### Analysis of the Impact of Oil Industry Employment

## Guiding question 5 : 
Is there a relationship between the oil industry employment rate and Calgary's unemployment Rate?

We begin by analyzing the impact of the oil industry on other sectors. Next, we merge the monthly Calgary employment data. After performing initial data cleaning, we integrate industry employment data with unemployment data. Finally, we visualize the relationship between oil industry employment and Calgary's unemployment rate.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
industries = "industries_cleaned.csv"
industries_data = pd.read_csv(industries)

# Convert 'Date' column to datetime format
industries_data['Date'] = pd.to_datetime(industries_data['Date'], format='%b-%y')

# Extract Year
industries_data['Year'] = industries_data['Date'].dt.year

# Convert relevant columns to numeric
convert_columns = ["Construction", "Forestry, fishing, mining, quarrying, oil and gas 10 11",
                   "Agriculture 9", "Finance, insurance, real estate, rental and leasing",
                   "Educational services"]
industries_data[convert_columns] = industries_data[convert_columns].replace(',', '', regex=True).astype(float)

# Group by Year & sum employment data
yearly_industries = industries_data.groupby('Year').agg({
    "Construction": "sum", 
    "Forestry, fishing, mining, quarrying, oil and gas 10 11": "sum",
    "Agriculture 9": "sum", 
    "Finance, insurance, real estate, rental and leasing": "sum",
    "Educational services": "sum"
}).reset_index()

# Rename the oil industry column for easier referencing
yearly_industries.rename(columns={"Forestry, fishing, mining, quarrying, oil and gas 10 11": "Oil Industry"}, inplace=True)

# Set Seaborn theme for better visuals
sns.set_style("whitegrid")
palette = sns.color_palette("husl", 5)

# Select industries to analyze
industries_to_analyze = ["Construction", "Finance, insurance, real estate, rental and leasing", "Agriculture 9", "Educational services"]

# Create scatter plots with trend lines
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, industry in enumerate(industries_to_analyze):
    row, col = divmod(idx, 2)
    
    # Scatter plot with trend line
    sns.regplot(data=yearly_industries, x="Oil Industry", y=industry, ax=axes[row, col], 
                scatter_kws={'s': 100, 'color': palette[idx], 'edgecolor': "black"},  # Scatter styling
                line_kws={'color': 'black', 'lw': 2},  # Trend line styling
                lowess=True)  # LOESS trend line (non-linear smoothing)

    # Title and labels
    axes[row, col].set_title(f"Oil Industry vs {industry}", fontsize=14, fontweight='bold')
    axes[row, col].set_xlabel("Oil Industry Employment", fontsize=12)
    axes[row, col].set_ylabel(f"{industry} Employment", fontsize=12)

# Adjust layout and show plots
plt.suptitle("Impact of Oil Industry Employment on Other Sectors", fontsize=16, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Analysis of the Impact of Oil Industry Jobs on Other Industries
The scatter plots represent the relationship of oil industry jobs with other industry jobs. Below is a summary analysis of the response of other industries to variations in oil industry jobs:

Oil Industry and Construction
There exists a positive correlation of oil industry jobs with construction jobs up to a point.
Construction jobs increase with oil sector jobs but levels off or decreases somewhat after reaching its peak.

Oil Industry vs. Finance, Insurance, Real Estate, Rental, and Leasing
The same trend is observed where this industry employment increases with oil sector employment but levels off or declines slightly at higher levels.
This shows that financial and real estate-related employment reaps the rewards of oil sector expansion, but only until a certain level.

Oil Industry and Agriculture
The relationship appears to be negative, in the sense that more oil industry employment comes after reduced employment in agriculture.
This could be due to workers moving from agriculture to the oil industry or financial conditions on farm production.
Oil Industry and Educational Services

There appears to be a positive relationship at first, but employment in educational services appears to level off or fall as oil employment levels increase.
This can translate into increased employment in the oil industry meaning an increased need for education positions, but only to that extent.
